```python
import matplotlib.pyplot as plt
import nashpy as nash
import numpy as np
```

# Replicator Dynamics

```python
A = np.array([
    [1, -1],
    [-1, 1]
])
road = nash.Game(A)
```

```python
xs_1 = road.replicator_dynamics()
```

```python
plt.figure(figsize=(13, 6))
plt.plot(xs_1)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.title('Replicator Dynamics (initial population: R = 0.5, L = 0.5)', fontsize=20)
plt.legend(['R', 'L'], fontsize=16)
plt.savefig('replicator_dynamics_example_1.png', transparent=True, dpi=300)
plt.show()
```

```python
xs_2 = road.replicator_dynamics(y0=[0.501, 0.499])
```

```python
plt.figure(figsize=(13, 6))
plt.plot(xs_2)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.title('Replicator Dynamics (initial population: R = 0.51, L = 0.49)', fontsize=20)
plt.legend(['R', 'L'], fontsize=16)
plt.savefig('replicator_dynamics_example_2.png', transparent=True, dpi=300)
plt.show()
```

```python
xs_3 = road.replicator_dynamics(y0=[0.499, 0.501])
```

```python
plt.figure(figsize=(13, 6))
plt.plot(xs_3)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.title('Replicator Dynamics (initial population: R = 0.49, L = 0.51)', fontsize=20)
plt.legend(['R', 'L'], fontsize=16)
plt.savefig('replicator_dynamics_example_3.png', transparent=True, dpi=300)
plt.show()
```

# Asymmetric Replicator Dynamics

```python
A = np.array([
    [0, 1, -1],
    [-1, 0, 1],
    [1, -1, 0]
])
B = np.array([
    [0, -1, 1],
    [1, 0, -1],
    [-1, 1, 0]
])
game = nash.Game(A, B)
```

```python
A = np.array([
    [1 ,5],
    [0, 3]
])
B = np.array([
    [1, 0],
    [5, 3]
])
game = nash.Game(A, B)
```

```python
xs, ys = game.asymmetric_replicator_dynamics(
    timepoints=np.linspace(0, 100, 100),
    # x0=np.array([0.1, 0.9]),
    # y0=np.array([0.001, 0.999])
)
```

```python
plt.figure(figsize=(13, 6))
plt.plot(xs)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.title('Asymmetric Replicator Dynamics', fontsize=20)
plt.legend(["Defect", "Cooperate"])
plt.show();
```

```python
plt.figure(figsize=(13, 6))
plt.plot(ys)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.title('Asymmetric Replicator Dynamics', fontsize=20)
plt.legend(["Defect", "Cooperate"])
plt.show();
```
